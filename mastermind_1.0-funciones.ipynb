```python
import random


# Function to print the mastermind board
def print_mastermind_board(CPU_choice, guess_codes, guess_flags):


    print('*'*50)
    print("          Richo's & Mike's MASTERMIND")
    print("*"*50)

    print("        |", end="")
    for x in CPU_choice:
        print("    " + x[:3], end="")
    print()	

    for i in range(len(guess_codes)):
        print("*"*50)

        print(hints[i][0], hints[i][1], hints[i][2], hints[i][3], end=" |")
        for x in guess_codes[i]:
            print("    " + x[:3], end="")

        print()	
    print("*"*50)

# The Main function
if __name__ == '__main__':

    # List of colors
    colors = ["RED", "GREEN", "YELLOW", "BLUE", "BLACK", "PINK"]

    # Mapping of colors to numbers	
    colors_code = {1:"RED", 2:"GREEN", 3:"YELLOW", 4:"BLUE", 5:"BLACK", 6:"PINK"}

    # Randomly selecting a passcode
    CPU_choice = random.sample(colors,4)
    
    # Number of chances for the player
    chances = 8

    # The passcode to be shown to the user
    show_passcode = ['???', '???', '???', '???']

    # The codes guessed by the player each turn
    guess_codes = [['---', '---', '---', '---'] for x in range(chances)]

    # The clues provided to the player each turn
    hints = [['-', '-', '-', '-'] for x in range(chances)]


    # The current turn
    turn = 0

    # The GAME LOOP
    while turn < chances:

        print("Select four non duplicate colors by entering a number from the below list.")
        print("1: Red, 2: Green, 3: Yellow, 4: Blue, 5: Black, 6: Pink")
        print("*"*50)
        print_mastermind_board(show_passcode, guess_codes, hints)

        # Accepting the player input 
        try:
            player_input = list(map(int, input("Enter your choice. Number between 1 and 6 separated by blank. No duplicates.").split()))
        
        # Check if integers where entered.    
        except ValueError:
            print("Error: Please enter a list of numbers.")
            continue   
 
        # Check if player entered 4 integers
        if len(player_input) != 4:
            print("Error: Please enter 4 numbers.")
            continue
    
        # Check if player entered duplicates
        if len(player_input) != len(set(player_input)):
            print("Error: Please avoid duplicates and re-insert your choice.")
            continue
         
        # Check if player entered integers from available list
        error = 0
        for x in player_input:
            if x > 6 or x < 1:
                error = 1
     
        if error == 1:           
            print("Error: Please enter a number from the provided list.")
            continue

        # Storing the player input
        for i in range(4):
            guess_codes[turn][i] = colors_code[player_input[i]]

        # Process to apply clues according to the player input
        temp_CPU_choice = [x for x in CPU_choice]

        pos = 0

        # Loop to set up clues for the player move
        for x in player_input:
            if colors_code[x] in temp_CPU_choice:
                if player_input.index(x) == CPU_choice.index(colors_code[x]):
                    hints[turn][pos] = 'R'
                else:
                    hints[turn][pos] = 'W'
                pos += 1
                temp_CPU_choice.remove(colors_code[x])

        random.shuffle(hints[turn])


        # Check for win condition
        if guess_codes[turn] == CPU_choice:
            print_mastermind_board(CPU_choice, guess_codes, hints)
            print("Congratulations!! YOU WIN!!!!")
            break

        # Update turn
        turn += 1

# Check for loss condiiton
if turn == chances:
    print_mastermind_board(CPU_choice, guess_codes, hints)
    print("YOU LOSE!!! Better luck next time!!!")	
```
